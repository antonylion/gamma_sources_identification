```python
data_path = '/var/tmp/zapp_an-a/Desktop/try/sim_gray_full_dataset'
output_path = '/var/tmp/zapp_an-a/Desktop/try/sim_gray_augmented_dataset'

# Create the output directory if it doesn't exist
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Iterate through the images in the data path
for img_name in os.listdir(data_path):
    print('d')
    img_full_path = os.path.join(data_path, img_name)
    output_img_path = os.path.join(output_path, img_name)
    
    # Read the image using OpenCV
    image = cv2.imread(img_full_path, cv2.IMREAD_GRAYSCALE)
    
    # Rotate the image by 90 degrees clockwise
    rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
    
    # Save the rotated image
    cv2.imwrite(output_img_path, rotated_image)
```

```python
input_labels_path = 'sources_pixels-2000.txt'
output_labels_path = 'augmented-sources_pixels-2000.txt'

# Read the original labels from the CSV file
original_labels = np.genfromtxt(input_labels_path, delimiter=',')

# Create the corresponding rotated labels
rotated_labels = []
for index, x, y in original_labels:
    rotated_x = original_labels.shape[1] - y  # Rotate x-coordinate
    rotated_y = x  # Rotate y-coordinate
    rotated_labels.append([int(2000 + int(index)), rotated_x, rotated_y])

# Save the rotated labels to the output CSV file
np.savetxt(output_labels_path, rotated_labels, delimiter=',', fmt='%1.8f')
```
